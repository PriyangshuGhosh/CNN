```python
import torch 
import torch.nn as nn

model = nn.Sequential(
    nn.Conv2d(3,8,3),
    nn.ReLU(),
    nn.Conv2d(8,16,3),
    nn.ReLU(),
    nn.Flatten(), 
    nn.Linear(12544,10)
)
x=torch.randn(3,32,32)
x=x.unsqueeze(0)

y=model(x)
print(y.shape)
```
