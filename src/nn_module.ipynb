```python
import torch
import torch.nn as nn

class ConvNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.feature=nn.Sequential(
            nn.Conv2d(1,32,3,padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2,2),
            nn.Conv2d(32,64,3),
            nn.ReLU(),
            nn.MaxPool2d(2,2)
        )
        self.classifier=nn.Sequential(
            nn.Flatten(),
            nn.Linear(64*6*6,128),
            nn.ReLU(),
            nn.Linear(128,10)
        )
    def forward(self,x):
        x=self.feature(x)
        logits = self.classifier(x)
        return logits

model=ConvNN()
x=torch.randn(1,1,28,28)
y=model(x)
z=model.forward(x)
print(y.shape)
```
