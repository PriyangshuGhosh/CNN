```python
import torch 
shape=(2,3)
tensor = torch.rand(shape)
print(tensor.dtype)
print(tensor.device)
print(tensor.shape)
```

slicing (Indexing)

```python
print(tensor)
print(tensor[:,0:3])# here 0:3 means 0,1,2 columns 
tensor = torch.ones(4, 4)

# Accessing data
print(f"First row: {tensor[0]}")
print(f"First column: {tensor[:, 0]}")
print(f"Last column: {tensor[..., -1]}")

# Modifying data in-place using slicing
tensor[:, 1] = 0 
# This sets every element in the second column (index 1) to zero
print(tensor)
```

concatenation is adding another tensor without changing the dimension whereas 
stacking is like creating another dimension

```python
x = torch.tensor([[1,2,3],[4,5,6]])
y=torch.tensor([[7,8,9],[10,11,12]])

a=torch.cat((x,y),dim=0) # concatenating along the rows
b=torch.stack((x,y),dim=2) # stacking along the columns
print(a)
print(a.shape)
print(b)
print(b.shape)
print(x.shape)
print(y.shape)
```

matrix multiplication is @ 
and 
element wise multiplication is *

```python
tensor2 = torch.rand(3,4)
tensor3 = torch.rand(3,4)

print(tensor@tensor2) # matrix multiplication
print(tensor2*tensor3) # elementwise multiplication
```

from numpy to torch

```python
import numpy as np 

arr=np.random.rand(3,4)
ten = torch.from_numpy(arr)
print(arr)
print(ten)
```

from another tensor

torch.ones_like(x_data): Creates a tensor filled with 1s using the same shape and type as x_data.

torch.rand_like(x_data, dtype=torch.float): Creates a tensor with random values, keeping the shape but overriding the data type to a float.

```python
tensor4 = torch.rand_like(ten,dtype=torch.float64)
tensor5 = torch.ones_like(ten,dtype=torch.int32)
print(tensor4)
print(tensor5)
```
